Hệ thống LMS đang chạy 2 khóa học chính là Python và C++. Sếp Giám đốc Đào tạo (Academic Director) yêu cầu bạn xuất một báo cáo đánh giá chất lượng các lớp học.

Sếp hỏi: "Ở từng Khóa học, và cụ thể là từng Lớp, hiện có bao nhiêu học viên, điểm trung bình là bao nhiêu, và điểm cao nhất lớp là mấy để anh đi trao thưởng?"

```python
import pandas as pd

data_lms = {
    'Khoa_Hoc': ['Python', 'Python', 'Python', 'Python', 'C++', 'C++', 'C++'],
    'Lop_Hoc': ['PY-Sáng', 'PY-Sáng', 'PY-Tối', 'PY-Tối', 'CPP-01', 'CPP-01', 'CPP-01'],
    'Hoc_Vien': ['An', 'Bình', 'Cường', 'Dũng', 'Em', 'Hà', 'Giang'],
    'Diem_Thi': [8.5, 9.0, 4.0, 7.5, 8.0, 5.5, 9.5],
    'So_Buoi_Vang': [0, 1, 4, 0, 2, 3, 0]
}
df_lms = pd.DataFrame(data_lms)

print("--- Dữ liệu thô hệ thống LMS ---")
display(df_lms)
```

```python
bao_cao_lms = df_lms.groupby(['Khoa_Hoc','Lop_Hoc']).agg(
    Si_So = ('Hoc_Vien','count'),
    Diem_Trung_Binh = ('Diem_Thi', 'mean'),
    Diem_Cao_Nhat = ('Diem_Thi', 'max')
).reset_index()
display(bao_cao_lms)
```
