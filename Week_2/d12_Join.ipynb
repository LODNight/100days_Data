```python
import pandas as pd

df_profile = pd.DataFrame({
    'ID': ['HV01', 'HV02', 'HV03', 'HV04'],
    'Ten': ['An', 'Bình', 'Cường', 'Dũng'],
    'Lop': ['Python', 'Python', 'C++', 'C++']
})

df_score = pd.DataFrame({
    'ID': ['HV01', 'HV02', 'HV03', 'HV05'],
    'Mon_Thi': ['Code', 'Code', 'Toan', 'Toan'],
    'Diem': [8.5, 9.0, 7.5, 10.0]
})
display(df_profile)
display(df_score)
```

```python

print("--- 1. INNER JOIN (Mặc định) ---")
df_inner = pd.merge(df_profile, df_score, on='ID', how='inner')
display(df_inner)
```

```python
print("\n--- 2. LEFT JOIN (Giữ trọn vẹn bảng Profile) ---")
df_left = pd.merge(df_profile, df_score, on='ID', how='left')
display(df_left)

```
