```python
# .iloc (Integer Location): Ch·ªçn theo V·ªä TR√ç S·ªê (gi·ªëng index trong C/List).
# V√≠ d·ª•: L·∫•y d√≤ng th·ª© 0, d√≤ng th·ª© 1. (M√°y t√≠nh hi·ªÉu).

# .loc (Label Location): Ch·ªçn theo NH√ÉN (T√äN).
# V√≠ d·ª•: L·∫•y d√≤ng c√≥ t√™n "Cafe Sua", l·∫•y c·ªôt "Gia_Ban". (Con ng∆∞·ªùi hi·ªÉu).

import pandas as pd
import io

menu_data = {
    'Ten_Mon': ['Cafe Den', 'Cafe Sua', 'Bac Xiu', 'Tra Dao', 'Sinh To Bo'],
    'Gia_Von': [10, 12, 15, 8, 20],
    'Gia_Ban': [20, 25, 35, 30, 50],
    'So_Luong_Ban': [50, 80, 40, 60, 20]
}

df_coffee = pd.DataFrame(menu_data)

# Tong_Lai = SL 1 ly * SL ban
df_coffee["Gia_1_Ly"] = df_coffee["Gia_Ban"] - df_coffee["Gia_Von"]

df_coffee["Tong_Lai"] = df_coffee["Gia_1_Ly"] * df_coffee["So_Luong_Ban"]


```

```python
# Gi·∫£ s·ª≠ df_coffee ƒëang c√≥ index m·∫∑c ƒë·ªãnh l√† 0, 1, 2, 3, 4
# ƒê·ªÉ d√πng .loc s∆∞·ªõng h∆°n, ta h√£y set c·ªôt 'Ten_Mon' l√†m Index (Nh√£n d√≤ng)

df_menu = df_coffee.set_index("Ten_Mon")
print("--- DataFrame ƒë√£ set Index ---")
display(df_menu)

# 1. D√πng .iloc (V·ªã tr√≠ s·ªë - D√†nh cho m√°y t√≠nh)
# C√∫ ph√°p: [H√†ng, C·ªôt]
print("\n--- .iloc (L·∫•y d√≤ng th·ª© 2: Bac Xiu) ---")
print(df_menu.iloc[2])

# 2. D√πng .loc (G·ªçi t√™n - D√†nh cho con ng∆∞·ªùi)
print("\n--- .loc (L·∫•y th√¥ng tin m√≥n 'Tra Dao') ---")
print(df_menu.loc["Tra Dao"])

# 3. L·∫•y ƒë√∫ng 1 √¥ c·ª• th·ªÉ (Sniper Shot üéØ)
# "Cho t√¥i bi·∫øt L·ª£i nhu·∫≠n (Tong_Lai) c·ªßa Cafe Sua"
lai_cafe_sua = df_menu.loc['Cafe Sua', 'Tong_Lai']
print(f"\nüí∞ L·ª£i nhu·∫≠n Cafe S·ªØa: {lai_cafe_sua}k")
```

```python
# D√πng .loc ƒë·ªÉ l·∫•y ra Gi√° B√°n v√† L·ª£i Nhu·∫≠n T·ªïng c·ªßa 2 m√≥n: 'Bac Xiu' v√† 'Sinh To Bo'.
gia_2_mon = df_menu.loc[['Bac Xiu','Sinh To Bo'], ['Gia_Ban','Tong_Lai']]
print(gia_2_mon)
```
