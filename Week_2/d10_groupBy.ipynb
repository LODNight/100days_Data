```python
import pandas as pd

data = {
    'Nhom_Mon': ['Ca Phe', 'Ca Phe', 'Ca Phe', 'Tra', 'Sinh To', 'Tra'],
    'Ten_Mon': ['Cafe Den', 'Cafe Sua', 'Bac Xiu', 'Tra Dao', 'Sinh To Bo', 'Tra Vai'],
    'So_Luong': [50, 80, 40, 60, 20, 50],
    'Tong_Lai': [500, 1040, 800, 1320, 600, 1000]
}

df = pd.DataFrame(data)

print("--- Dữ liệu gốc ---")
display(df)


```

```python
print("--- 1. GROUP BY Cơ bản ---")
# Cú pháp: df.groupby('Cột_để_chia_nhóm')['Cột_để_tính_toán'].hàm_tính()
loi_nhuan_theo_nhom = df.groupby('Nhom_Mon')['Tong_Lai'].sum()

display(loi_nhuan_theo_nhom)

```

```python

print("--- 2. RESET_INDEX()  ---")
# Để kết quả biến thành 1 DataFrame chuẩn đẹp (dễ vẽ biểu đồ sau này), ta thêm .reset_index()
bao_cao_chuan = df.groupby('Nhom_Mon')['Tong_Lai'].sum().reset_index()
display(bao_cao_chuan)
```

```python
# Đề bài: Sếp thấy nhóm Trà và Cà Phê bán rất chạy. Sếp muốn biết: Trung bình mỗi món trong từng nhóm bán được bao nhiêu ly? (Để xem có nên thêm món mới vào nhóm đó không).

tb_so_ly = df.groupby('Nhom_Mon')['So_Luong'].mean().reset_index()
display(tb_so_ly)
display(tb_so_ly)
```
