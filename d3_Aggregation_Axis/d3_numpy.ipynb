```python
import numpy as np

# Tạo dữ liệu
# - np.array(): Chuyển list thường thành NumPy array.
# - np.arange(start, stop, step): Tạo dãy số (giống vòng lặp for i in range).
# - np.random: Tạo dữ liệu ngẫu nhiên (cực quan trọng để test khi chưa có data thật).

# 1. Chuyển đổi dữ liệu thô thành mảng để xử lý
raw_scores = [7.5, 8.0, 6.7, 9.1]
arr_scores = np.array(raw_scores)
print("Dữ liệu thô",arr_scores)

# 2. Tạo một cột số thứ tự (ID) cho 100 nhân viên
employee_id = np.arange(1,101)
print(employee_id)

# 3. Giả lập dữ liệu doanh thu của 5 cửa hàng trong 3 ngày (Ma trận 3x5)
# randint(low, high, size=(rows, cols))
sales_data = np.random.randint(100, 500, size=(3,5))
print(sales_data)

```
