```python

# Bối cảnh: Bạn có dữ liệu điểm của 5 học viên.
# Điểm thi (NumPy Array): Gồm 3 cột [Chuyên cần, Giữa kỳ, Cuối kỳ].
# Điểm cộng (Pandas Series): Một số bạn có đóng góp xây dựng bài nên được cộng điểm thưởng, nhưng không phải ai cũng có.

# Nhiệm vụ:
# Tính điểm tổng kết theo trọng số: 10% - 30% - 60%.
# Cộng điểm thưởng vào điểm tổng kết (Cẩn thận dữ liệu thiếu).
# Lọc ra danh sách những bạn "Cần hỗ trợ" (Điểm tổng kết < 5.0).

import pandas as pd
import numpy as np

# Danh sách học viên
students = ['An', 'Bình', 'Cường', 'Dũng', 'Em']

# 1. Bảng điểm thô (NumPy Array) - [CC, GK, CK]
# Lưu ý: Cường bị 0 điểm cuối kỳ (ví dụ vắng thi)
scores_raw = np.array([
    [9, 8, 8],   # An
    [6, 5, 5],   # Bình
    [8, 7, 0],   # Cường
    [9, 9, 9],   # Dũng
    [5, 4, 4]    # Em
])

# 2. Trọng số (Weights)
weights = np.array([0.1, 0.3, 0.6])

# 3. Điểm cộng (Pandas Series) - Chỉ 3 bạn có
bonus_points = pd.Series([0.5, 1.0, 0.2], index=['An', 'Dũng', 'Em'])

print("Dữ liệu đã sẵn sàng!")
```

```python
# ============== Task 1 =============== 
# Tính điểm tổng kết theo trọng số: 10% - 30% - 60%.
# Dùng dot product (nhân ma trận) giữa scores_raw và weights.
final_scores = scores_raw.dot(weights)

print("-"*30)
print("1. Điểm tổng kết\n", final_scores)


# ============== Task 2 =============== 
# Cộng điểm thưởng vào điểm tổng kết (Cẩn thận dữ liệu thiếu).
# Biến kết quả Task 1 thành Pandas Series (để có tên học viên), sau đó cộng với bonus_points. 

# Kết nối bảng final_scores với students
total_scores = pd.Series(data = final_scores, index = students)
# Tính phần giao (các bạn đc bonus), nếu các bạn ko đc bonus thì sẽ + 0 
total_scores = total_scores.add(bonus_points, fill_value = 0)
# Tránh cộng lố (0-10)
total_scores = np.clip(total_scores, 0, 10)

print("-"*30)
print("2. Điểm sau bonus\n", total_scores)



# ============== Task 3 =============== 
# Boolean Indexing (Filtering)
# Lọc ra danh sách những bạn "Cần hỗ trợ" (Điểm tổng kết < 5.0).
print("-"*30)
at_risk_students = total_scores[total_scores < 5.0]
print("3. Các bạn cần Hỗ trợ thêm\n", at_risk_students)


```
