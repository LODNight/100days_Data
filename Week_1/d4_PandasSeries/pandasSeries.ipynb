```python
import pandas as pd
import numpy as np

# 1. Tạo Series từ List (Index mặc định là 0, 1, 2...)
diem_so = [8.5, 9.0, 7.5, np.nan] # 1 ban vang thi

s1 = pd.Series(diem_so)
print(f"Series mac dinh:\n{s1}")
print("-" * 20)


# 2. Tạo Series có Index (Đây là cách dùng chuẩn)
# Giả lập dữ liệu LMS: Tên học viên gắn liền với điểm

hoc_vien = ['An', 'Bình', 'Cường', 'Dũng']
diem = [8.5, 9.0, 7.5, 6.0]
s_lms = pd.Series(data = diem, index = hoc_vien)
print("Series\n",s_lms)

print("-" * 20)
print("Diem cua ban Binh:", s_lms['Bình'])
s_lms.name = "Điểm của các bạn trên 8" # bổ sung thêm name cho s_lms
print("Các bạn có điểm trên 8:\n",s_lms[s_lms > 8]) # lấyt data so sánh
```

```python
# Đề bài: Bạn có điểm thi của lớp A và lớp B (lưu ý tên học viên có thể trùng hoặc khác nhau).

# Lớp A: {'An': 8, 'Bình': 9}
# Lớp B: {'An': 7, 'Cường': 10}

# Yêu cầu: Tính điểm trung bình của từng học viên qua 2 lớp (Cộng lại chia đôi).

# Gợi ý: Chỉ cần cộng trực tiếp series_a + series_b. Hãy xem điều gì xảy ra với bạn 'Bình' và 'Cường' 
# (những người chỉ học 1 lớp).

lop_a = pd.Series({'An': 8, 'Bình': 9})
lop_b = pd.Series({'An': 7, 'Cường': 10})

tong_ket = (lop_a + lop_b) / 2      # Pandas lấy index ra để làm phép tính, nếu khớp thì sẽ cộng lại
print("Ket qua Tong Ket:\n", tong_ket)

# Sử dụng cách chuyên nghiệp hơn để tránh lỗi NaN
tong_chuan = lop_a.add(lop_b, fill_value=0) / 2
print("Ket qua Tong Chuan:\n", tong_chuan)
```
