```python
# Bối cảnh: Bạn là chủ quán cà phê "Code & Coffee". Cuối ngày, bạn có số liệu bán hàng của 5 món đồ uống chủ đạo. Bạn cần tính toán xem món nào đang "gánh team" về mặt lợi nhuận, món nào bán chạy nhưng lãi thấp.

# Khởi tạo: Tạo DataFrame df_coffee từ dictionary trên.
# Doanh thu (Revenue): Tạo cột Doanh_Thu = Gia_Ban * So_Luong_Ban.
# Lợi nhuận (Profit):
# Tính lãi trên mỗi ly: Lai_1_Ly = Gia_Ban - Gia_Von.
# Tính tổng lợi nhuận: Tong_Lai = Lai_1_Ly * So_Luong_Ban.
# Báo cáo (Inspection):
# Dùng .info() để kiểm tra xem có cột nào bị lỗi không.
# (Nâng cao): In ra món có Tong_Lai cao nhất.

import pandas as pd
import io

menu_data = {
    'Ten_Mon': ['Cafe Den', 'Cafe Sua', 'Bac Xiu', 'Tra Dao', 'Sinh To Bo'],
    'Gia_Von': [10, 12, 15, 8, 20],   # Giá vốn (ngàn đồng)
    'Gia_Ban': [20, 25, 35, 30, 50],  # Giá bán (ngàn đồng)
    'So_Luong_Ban': [50, 80, 40, 60, 20] # Ly
}

df = pd.DataFrame(menu_data)

print("==== Danh sách món ====")
display(df)

print("==== Doanh thu ====")
df["Doanh_Thu"] = df["Gia_Ban"] * df["So_Luong_Ban"]

df["Lai_1_Ly"] = df["Gia_Ban"] - df["Gia_Von"]
df["Tong_Lai"] = df["Lai_1_Ly"] * df["So_Luong_Ban"]
display(df)

print("\n=========== Check =============")
df.info()

print("\n=========== BEST SELL =============")
top_profit = df.sort_values("Tong_Lai", ascending=False).iloc[0]
print(f"\n TOP: {top_profit["Ten_Mon"]} (Lãi: {top_profit["Tong_Lai"]})")
```
