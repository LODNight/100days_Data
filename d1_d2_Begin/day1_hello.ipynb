```python
import pandas as pd
import numpy as np

# 1. T·∫°o d·ªØ li·ªáu gi·∫£ (Thay v√¨ d√πng struct trong C hay Class trong Python App)
data = {
    'Ten_Hoc_Vien': ['An', 'Binh', 'Cuong', 'Dung', 'Em', 'Ha'],
    'Tuoi': [12, 13, 12, 14, 13, 12],
    'Diem_Python': [8.5, 9.0, np.nan, 7.5, 6.0, 9.5],  # np.nan l√† gi·∫£ l·∫≠p d·ªØ li·ªáu b·ªã thi·∫øu (v√≠ d·ª•: v·∫Øng thi)
    'Diem_C': [7.0, 8.0, 6.5, np.nan, 9.0, 8.5],
    'Diem_D': [8.2, 9.5, 7.1, np.nan, 9.0, 8.5]
}

# 2. Chuy·ªÉn dictionary th√†nh DataFrame (B·∫£ng d·ªØ li·ªáu)
# ƒê√¢y l√† ƒë·ªëi t∆∞·ª£ng quan tr·ªçng nh·∫•t b·∫°n s·∫Ω l√†m vi·ªác trong 100 ng√†y t·ªõi
df = pd.to_series = pd.DataFrame(data)

# 3. Xem d·ªØ li·ªáu
print("D·ªØ li·ªáu th√¥:")
display(df) # Trong Jupyter, d√πng display() s·∫Ω ƒë·∫πp h∆°n print()

# 4. Th·ª≠ t√≠nh to√°n nhanh (Data Analysis c∆° b·∫£n)
print("\n--- Th·ªëng k√™ nhanh ---")
print(f"ƒêi·ªÉm Python trung b√¨nh: {df['Diem_Python'].mean()}")
print(f"H·ªçc vi√™n l·ªõn tu·ªïi nh·∫•t: {df['Tuoi'].max()}")
```

```python
# Cell 2: H·ªèi d·ªØ li·ªáu v√†i c√¢u ƒë∆°n gi·∫£n

# C√¢u 1: ƒêi·ªÉm Python trung b√¨nh c·ªßa l·ªõp l√† bao nhi√™u?
avg_python = df['Diem_Python'].mean()

# C√¢u 2: Ai l√† ng∆∞·ªùi l·ªõn tu·ªïi nh·∫•t?
max_age = df['Tuoi'].max()

# C√¢u 3: D·ªØ li·ªáu ƒëang c√≥ v·∫•n ƒë·ªÅ g√¨ kh√¥ng? (G√≥c nh√¨n Senior)
missing_count = df['Diem_Python'].isnull().sum()

print(f"üîπ ƒêi·ªÉm trung b√¨nh Python: {avg_python}")
print(f"üîπ Tu·ªïi l·ªõn nh·∫•t: {max_age}")
print(f"‚ö†Ô∏è S·ªë b·∫°n ch∆∞a c√≥ ƒëi·ªÉm Python (b·ªã thi·∫øu): {missing_count}")
```

```python
import numpy as np
import time

# --- PH·∫¶N 1: T·∫†O M·∫¢NG (ARRAY) ---
# List Python th√¥ng th∆∞·ªùng
python_list = [1, 2, 3, 4, 5]

# NumPy Array (Nh√¨n th√¨ gi·ªëng list, nh∆∞ng b·∫£n ch·∫•t l√† C Array)
numpy_arr = np.array([1, 2, 3, 4, 5])

print(f"List th∆∞·ªùng: {python_list}")
print(f"NumPy Array: {numpy_arr}")
print("-" * 30)

# --- PH·∫¶N 2: BROADCASTING (T√çNH TO√ÅN KH√îNG LOOP) ---
# B√†i to√°n: C·ªông th√™m 10 v√†o t·∫•t c·∫£ ph·∫ßn t·ª≠

# C√°ch Python List (Ph·∫£i d√πng Loop ho·∫∑c Map)
# [x + 10 for x in python_list]

# C√°ch NumPy (Th·∫≥ng tay c·ªông nh∆∞ s·ªë h·ªçc) -> ƒê√ÇY L√Ä VECTORIZATION
result = numpy_arr + 10 

print(f"K·∫øt qu·∫£ c·ªông 10: {result}")
print("-" * 30)

# --- PH·∫¶N 3: TH·ª¨ TH√ÅCH T·ªêC ƒê·ªò (BENCHMARK) ---
# T·∫°o 1 tri·ªáu ph·∫ßn t·ª≠ ng·∫´u nhi√™n
size = 1_000_000
large_list = list(range(size))
large_array = np.array(range(size))

# ƒêo t·ªëc ƒë·ªô List Python
start = time.time()
large_list_squared = [x**2 for x in large_list]
print(f"Python List t·ªën: {time.time() - start:.5f} gi√¢y")

# ƒêo t·ªëc ƒë·ªô NumPy
start = time.time()
large_array_squared = large_array ** 2
print(f"NumPy Array t·ªën: {time.time() - start:.5f} gi√¢y")
```

```python
# Gi·∫£ l·∫≠p ƒëi·ªÉm c·ªßa 3 h·ªçc vi√™n (H√†ng), m·ªói ng∆∞·ªùi c√≥ 3 c·ªôt ƒëi·ªÉm
# H·ªçc vi√™n A: 8, 9, 7
# H·ªçc vi√™n B: 5, 6, 5
# H·ªçc vi√™n C: 9, 10, 9
diem_tho = np.array([
    [8, 9, 7],
    [5, 6, 5],
    [9, 10, 9]
])

# Tr·ªçng s·ªë: Chuy√™n c·∫ßn 10%, Gi·ªØa k·ª≥ 30%, Cu·ªëi k·ª≥ 60%
trong_so = np.array([0.1, 0.3, 0.6])

# T√çNH ƒêI·ªÇM T·ªîNG K·∫æT (Ph√©p nh√¢n ma tr·∫≠n - Dot Product)
# Thay v√¨ vi·∫øt 2 v√≤ng l·∫∑p l·ªìng nhau (Nested Loop), ta d√πng np.dot
diem_tong_ket = diem_tho.dot(trong_so)
diem_tong_ket_lon_hon_8 = diem_tong_ket[diem_tong_ket >= 8]
diem_cao_nhat = diem_tong_ket.max()

print("B·∫£ng ƒëi·ªÉm th√¥ (Matrix):\n", diem_tho)
print("Tr·ªçng s·ªë:", trong_so)
print("-" * 20)
print("üëâ ƒêi·ªÉm t·ªïng k·∫øt (ƒê√£ nh√¢n tr·ªçng s·ªë):", diem_tong_ket)
print("üëâ ƒêi·ªÉm t·ªïng k·∫øt l·ªõn h∆°n 8:", diem_tong_ket_lon_hon_8)
print("üëâ ƒêi·ªÉm t·ªïng k·∫øt Cao nh·∫•t:", diem_cao_nhat)
```
