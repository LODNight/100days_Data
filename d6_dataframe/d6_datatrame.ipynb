```python
import pandas as pd
import io
# CÁCH 1: Tạo từ Dictionary (Thường dùng để test nhanh)
data_dict = {
    'Ten': ['An', 'Bình', 'Cường', 'Dũng', 'Em'],
    'Bo_Phan': ['Sale', 'IT', 'Sale', 'HR', 'IT'],
    'Luong': [1500, 2000, 1600, 1200, 1800], # Đơn vị: $
    'Tham_Nien': [2, 5, 3, 1, 4] # Đơn vị: Năm
}

df = pd.DataFrame(data_dict)

print("--- DataFrame từ Dictionary ---")
display(df)

# CÁCH 2: Đọc từ file CSV (Mô phỏng)
# Thực tế bạn sẽ dùng: df = pd.read_csv('duong_dan_file.csv')
csv_content = """ID,Ten,Diem
SV01,Lan,8.5
SV02,Mai,9.0
SV03,Hung,7.5"""

df_csv = pd.read_csv(io.StringIO(csv_content))
print("\n--- DataFrame từ CSV (Mô phỏng) ---")
display(df_csv)
```

```python
total = df["Luong"].mean()
print(total)
```

```python
# Đề bài: Sếp yêu cầu tính "Lương Thưởng Tết".

# Công thức: Lương * Hệ số thâm niên.

# Hệ số: Mỗi năm thâm niên được thưởng thêm 10% lương (tức là 0.1 * Tham_Nien).

# Gợi ý: Hãy tạo cột mới df['Thuong'] bằng cách nhân các cột cũ.

print("=== Lương Thưởng Tết ====")

# 1. Tính hệ số thưởng (Ví dụ 2 năm -> 0.2)
he_so = df["Tham_Nien"] * 0.1

# 2. Tính tiền thưởng
df["Thuong_Tet"] = df["Luong"] * he_so

df["Tong_Thu_Nhap"] = df["Luong"] + df["Thuong_Tet"]
display(df)

```
